```python
import quandl
import matplotlib.pyplot as plt
with open("key.txt","r") as f:
    x=f.read()
```

```python
ARM_TOT=quandl.get('WORLDBANK/ARM_TOT',authtoken=x)
LUX_TOT=quandl.get('WORLDBANK/LUX_TOT',authtoken=x)
```

```python
plt.plot(ARM_TOT)
plt.plot(LUX_TOT)
plt.show()
```

```python
data=ARM_TOT.sort_values(by='Value', ascending=0)
highest_second=data['Value'][1]
print 'Second higest value for Armenia is: ' + str(highest_second)
```

```python
import googlemaps
with open("google_key.txt","r") as f:
    x=f.read()
maps=googlemaps.Client(key=x)
```

```python
distance=[]
city=["Philadelphia, PA","Baltimore, MD","Washington, DC","San Francisco, CA",
           "Boston, MA","Los Angeles, CA", "Springfield, MA","Chicago, IL","Atlantic city, NJ", "Richmond, VA"]
for i in city:
    directions=maps.directions(i,"New York, NY",units="metric") 
    distance.append(directions[0]['legs'][0]['distance'])
for i in range(0,len(distance)):
    print('The distance between ' +  city[i] + ' and ' + " New York, NY" + ' is ' + distance[i]["text"])
```
